```python
import os
```

```python
import pandas as pd
```

```python
data=pd.read_csv("winequality-red.csv")
```

```python
data.sample(10)
```

```python
from sklearn.model_selection import train_test_split 
xtrain,xtest,ytrain,ytest=train_test_split(data.iloc[:,:-1],data.iloc[:,-1])
```

```python
xtrain
```

```python
from sklearn.linear_model import ElasticNet
lr=ElasticNet(alpha=0.1,l1_ratio=0.5,random_state=42)
```

```python
lr.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
import numpy as np
```

```python
def eval_metrics(actual,predict):
    rmse=np.sqrt(mean_squared_error(actual,predict))
    r2s=r2_score(actual,predict)
    mse=mean_absolute_error(actual,predict)

    return rmse,r2s,mse

predicted=lr.predict(xtest)
(rmse,r2s,mse)=eval_metrics(ytest,predicted)
```

```python
print("rmse:",rmse)
print("r2s:",r2s)
print("mse:",mse)
```
